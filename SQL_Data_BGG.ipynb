# Load Data from Database Using SQL

```python
import pandas as pd

from sqlalchemy import create_engine
```

```python
# Connect to sqlite db
db_file = r'datasets/database.sqlite'
engine = create_engine(r"sqlite:///{}".format(db_file))
```

```python
#find out what tables are in the database
sql = "SELECT name FROM sqlite_master WHERE type = 'table';"

table_names = pd.read_sql(sql, engine)
table_names
```

```python
#get all the rows from the BoardGames table in the database
sql = "SELECT * FROM BoardGames;"
grade_df = pd.read_sql(sql, engine)
grade_df.head()
```

```python
#get all the rows from the [bgg.ldaOut.topics] table in the database
sql = "SELECT * FROM [bgg.ldaOut.topics]"
grade_df_1 = pd.read_sql(sql, engine)
grade_df_1.head()
```

```python
# [ ] works for table 
sql = "SELECT * FROM [bgg.ldaOut.top.terms]"
grade_df_2 = pd.read_sql(sql, engine)
grade_df_2.head()
```

```python
# [ ] works for table
sql = "SELECT * FROM [bgg.ldaOut.top.documents]"
grade_df_3 = pd.read_sql(sql, engine)
grade_df_3.head()
```

```python
#' ' works for colume; [ ] works for table
sql = "SELECT 'document','topic' FROM [bgg.ldaOut.top.documents]"
grade_df_3a = pd.read_sql(sql, engine)
grade_df_3a.head()
```

```python
#double "" works for table

sql = 'SELECT * FROM "bgg.topics"'
grade_df_4 = pd.read_sql(sql, engine)
grade_df_4.head()
```

```python
#[ ] works for colume; double "" works for table

sql = 'SELECT [game.id], [details.maxplayers], [details.maxplaytime] FROM "bgg.topics"'
grade_df_4a = pd.read_sql(sql, engine)
grade_df_4a.head()
```

```python
# [ ] works for table
sql = 'SELECT * FROM [bgg.topics]'
grade_df_4 = pd.read_sql(sql, engine)
grade_df_4.head()
```

```python
# " " works for colume; [ ] works for table
sql = 'SELECT "game.id", "game.type", "details.maxplaytime" FROM [bgg.topics]'
grade_df_4 = pd.read_sql(sql, engine)
grade_df_4.head()
```
